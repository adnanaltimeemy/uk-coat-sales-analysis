```python
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer

# Load dataset
df = pd.read_csv("reviews.csv")

# Remove accidental spaces in column names
df.columns = df.columns.str.strip()

# Show column names
print("Columns:", df.columns)

# Correct column name
text_column = "text"

# Fill missing text values
df[text_column] = df[text_column].fillna("")

# Create TF-IDF vectorizer
tfidf = TfidfVectorizer(
    stop_words="english",
    max_features=100
)

# Transform the text column
X_text = tfidf.fit_transform(df[text_column])

# Show result shape
print("TF-IDF Matrix Shape:", X_text.shape)

# Convert to DataFrame for easier viewing
tfidf_df = pd.DataFrame(
    X_text.toarray(),
    columns=tfidf.get_feature_names_out()
)

print(tfidf_df.head())
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report
from sklearn.feature_extraction.text import TfidfVectorizer

df.columns = df.columns.str.strip()

df["purchased_again"] = (df["rating"] >= 4).astype(int)

vectorizer = TfidfVectorizer(stop_words="english")

X_text = vectorizer.fit_transform(df["text"])
y = df["purchased_again"]

X_train_text, X_test_text, y_train_text, y_test_text = train_test_split(
    X_text, y, test_size=0.3, random_state=42
)

nlp_model = LogisticRegression(max_iter=1000, class_weight="balanced")

nlp_model.fit(X_train_text, y_train_text)

y_pred_text = nlp_model.predict(X_test_text)

print(classification_report(y_test_text, y_pred_text, zero_division=0))
```
